```python
#ans1)
#plt.plot(data['Date'], data['Low'],color="red",label="Low")
df1['Open'].plot(label='Open', figsize=(15,7))
df1['Low'].plot(label='Low', figsize=(15,7))
plt.xlabel("Date")
plt.ylabel("Prices")
plt.title("Stock Price")
plt.legend()
plt.show()
```

```python
#ans2)
mean = (df1['High'] + df1['Low'] + df1['Open'] + df1['Close'])/4
a1 = (df1['High']-mean)*(df1['High']-mean)
a2 = (df1['Low']-mean)*(df1['Low']-mean)
a3 = (df1['Open']-mean)*(df1['Open']-mean)
a4 = (df1['Close']-mean)*(df1['Close']-mean)
b = (a1+a2+a3+a4)/4
std = b**0.5
# plt.plot(data['Date'], std)
std.plot(figsize=(15,7))
plt.xlabel("Date")
plt.ylabel("Deviated by")
plt.title("Stock Price")
plt.show()
```



* Standard Deviation = root(sum(xi-mean)2/n)
* It gives how much price deviates or changes within a day using high, low, open, close values.
*Then plot the graph of all dates.



```python
#ans3)
med = [df1['High'], df1['Low'], df1['Open'], df1['Close']]
df = pd.DataFrame(med)
plt.plot(df.median(),df1['Volume'])
plt.xlabel("Median")
plt.ylabel("Volume")
plt.title("Stock")
plt.yticks([360100, 1007500, 2206300, 3354800, 4937300])
plt.show()
```

1.Median of High, Low, Open and Close is identify by formula median().
2.The value of median is plotted with respect to volume of stocks.
3.No particular trend is observed relating to median and volume of stocks.

```python
mean = (df1['Volume'].sum()/df1['Date'].count() + df1['Volume'] - df1['Volume'])
mean.plot(figsize=(15,7), label='Mean')
# plt.plot(data['Date'], mean, label='Mean')
# plt.plot(data['Date'], data['Volume'],label="Original")
df1['Volume'].plot(label='Original', figsize=(15,7))
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Comparison between mean and original volume")
plt.yticks([139100, 360100, 728668, 1507500, 2206300, 3354800, 4937300])
plt.legend()
plt.show()
```

```python
#ans5)
df1['Stock Trading'].plot(label='Stock Trading', figsize=(15,7))
(df1['Open']-df1['Close']).plot(label='Open-Close', figsize=(15,7))
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Stock")
plt.legend()
plt.show()
```
