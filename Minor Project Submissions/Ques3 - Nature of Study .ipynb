```python
From given database we can see that there are two types of clinical studies, that is Interventional and Observational Study.
Interventional Study :- Here, participants are assigned to groups that receive one or more intervention/treatment (or no intervention) 
                        so that researchers can evaluate the effects of the interventions on biomedical or health-related outcomes. 
                        Participants may receive diagnostic, therapeutic, or other types of interventions.
Observational Study :- Here, participants are evaluated on the basis of biomedical or health outcomes. 
Observational [Patient Registry] :- Participants may receive diagnostic, therapeutic, or other types of interventions
                                    but the investigator does not assign participants to a specific interventions/treatment. 
Expanded access :- Considering the Patients with serious diseases or conditions who cannot participate in a clinical trial 
                   to gain access to a medical product.
```

```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
df = pd.read_csv("COVID clinical trials.csv")
```

```python
# Method 1 - Plotting the data using pie chart method
sizes = [3322, 2097, 330, 34]
labels = 'Interventional', 'Observational', 'Observational (Patient registry)', 'Expanded access'
colors = ("cyan", "brown", "yellow", "red")
textprops = {"fontsize" : 20.0}
plt.pie(sizes, labels = labels, colors = colors, autopct ='%1.0f%%')
plt.title('Nature of Study', fontsize = 25)
plt.axis('equal')
plt.show()
```

```python
# Method 2 - Plot the data using bar chart method
sizes = [3322, 2097, 330, 34]
labels = 'Interventional', 'Observational', 'Observational (Patient registry)', 'Expanded access'
colors = ("cyan", "brown", "yellow", "red")
plt.bar(labels, sizes, width= 0.5, color = colors, edgecolor="red")
plt.rcParams["figure.figsize"] = (16,11.5)
plt.title("Nature of Study", fontsize = 25)
plt.xlabel("Type of Study", fontsize = 15)
plt.ylabel("Count", fontsize = 15)
plt.show()
```
