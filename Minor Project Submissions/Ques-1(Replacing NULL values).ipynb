```python
#remove unwanted data
df.dropna()
```

```python
Now, to replace the NULL VALUES, four methods have been shown below:
FORWARD FILL,
BACKWARD FILL,
Replace by ZERO and 
INTERPOLATE.
```

```python
# Method-1 : forward fill to replace null values
df.fillna(method='ffill')
```

```python
# Method-2 : backward fill to replace null values
df.fillna(method='bfill')
```

```python
# Method-3 : replacing null values with ZERO
df.fillna(0)
```

```python
# Method-4 : interpolation to replace null values
import pandas as pd
df = pd.read_csv("COVID clinical trials.csv")
df.interpolate()
```

```python
Among all of the above methods of replacing null values, Method-3 is the most preferred choice.
The reason for the same is that we can't figure out the exact value of the given data in the remaining other methods.
Let's suppose, if, in future, we get some information about the data for the corresponding null value,
    then we won't be able to make changes.
    The reason for that is we won't be able to differentiate among :
        the actual data and the modified data obtained after replacing NULL values.
```
