# Question2
â€¢ Analyze the price change within a day using standard deviation.

# Answer 

```python
mean = (stock['High'] + stock['Low'] + stock['Close'])/4
a1 = (stock['High']-mean)*(stock['High']-mean)
a2 = (stock['Low']-mean)*(stock['Low']-mean)
a3 = (stock['Open']-mean)*(stock['Open']-mean)
a4 = (stock['Close']-mean)*(stock['Close']-mean)
b = (a1+a2+a3+a4)/4
std = b**0.5

std.plot(figsize=(15,7),color='green')
plt.xlabel("Date")
plt.ylabel("Deviated by")
plt.title("Stock Price")
plt.show()
```
