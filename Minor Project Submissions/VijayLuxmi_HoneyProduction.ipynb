Q4 Trend in total production and value of production 

```python
import pandas as pd
from matplotlib import pyplot as plt

# Read CSV into pandas
data = pd.read_csv(r"/content/HoneyProduction.csv")
data.head()
df = pd.DataFrame(data)
totalprod = df['totalprod']
# line-1
lst_prod = []
for i in totalprod:
  lst_prod.append(i)
lst_year = []
for j in df['year']:
  lst_year.append(j)
y = lst_prod
x = lst_year
plt.plot(x,y,label = "total_poduction over the years")
#plt.show()
# line chart 2
plt.legend()
x1 = df['year']
y1 = df['prodvalue']
plt.plot(x1,y1,label = "production value over the years")
plt.show()
```

Conclusion : here we can see over the years the total production has decreased but the value of production has increased .
As the production of honey has decreased hence the demand of it also would be increased .
