```python
# Distribution graphs (histogram/bar graph) of column data
def Distribution(df, nGraphShown, nGraphPerRow):
    nunique = df.nunique()
    df = df[[col for col in df if nunique[col] > 1 and nunique[col] < 50]]

    nRow, nCol = df.shape
    col_name = list(df)
    nGraphRow = (nCol + nGraphPerRow - 1) / nGraphPerRow
    plt.figure(num = None, figsize = (6 * nGraphPerRow, 8 * nGraphRow), dpi = 80, facecolor = 'w', edgecolor = 'k')
    for i in range(min(nCol, nGraphShown)):
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df.iloc[:, i]
        if (not np.issubdtype(type(columnDf.iloc[0]), np.number)):
            valueC = columnDf.value_counts()
            valueC.plot.bar()
        else:
            columnDf.hist()
        plt.ylabel('counts')
        plt.xticks(rotation = 90)
        plt.title(f'{col_name[i]} (column {i})')
        plt.tight_layout(pad = 2.0, w_pad = 2.0, h_pad = 1.5)
    plt.show()
Distribution(df,11,7)
```

## Q4) Use a map (world map) to show where these studies are taking place .

```python
def split(l):
    return l.split(',')[-1].strip()
df['Loc'] = df.Locations.apply(lambda x:split(str(x)))
```

```python
fig = go.Figure([go.Choropleth(
    locations=df.groupby(['Loc']).agg('count')['Rank'].index,
    z=df.groupby(['Loc']).agg('count')['Rank'].values.astype(float),
    locationmode='country names',
    colorscale='Reds',
    autocolorscale=False,
    marker_line_color='yellow',
    showscale = True,
)])
fig.update_layout(title='Study Locations')
fig.show()
```

# Q5) Analyze what age bracket and gender these studies are considering and different graphs to show these. Write your observations.

```python
import re
```

```python
df['Gender'].unique()
```

```python
def clean(age):
    if len(re.findall(r'\(.*\)',age)):
        return re.findall(r'\(.*\)',age)[0]
    return '('+age+')'
```

```python
Data_age = df.Age.apply(lambda x : clean(x))
plt.xticks(rotation=45)
Data_age.hist(color = 'red')
plt.grid(False)
```

```python
genderCount = df.groupby('Gender').count()['Rank']
df['Gender'].hist()
plt.grid(False)
```

```python
i = 0
fig = make_subplots(rows=3, cols=2, subplot_titles=list(pd.DataFrame(df.groupby(['AgeBrackets'])['Gender'].value_counts()).unstack().index))
for row in range(1,4):
    for col in range(1,3):
        dt = pd.DataFrame(df.groupby(['AgeBrackets'])['Gender'].value_counts()).unstack().iloc[i]
        fig.add_trace(go.Bar(x=dt.Gender.index,y=dt.Gender.values),row = row, col = col)        
        i+=1
fig.show()
```
