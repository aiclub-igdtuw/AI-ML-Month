```python
For Data Visualisation, we're using three methods to represent our data:
Method-1 : Through Pie Chart
Method-2 : Through Bar Graph
Method-3 : Through Histogram
```

```python
# Method 1 - Plot the data using pie chart method

sizes = [2805, 1025, 1004, 526, 181, 107, 74]
labels = 'Recruiting', 'Completed', 'Not yet recruiting', 'Active, not recruiting', 'Enrolling by invitation', 'Withdrawn', 'Terminated'
colors = ( "orange", "cyan", "brown", "grey", "indigo", "yellow", "red")
textprops = {"fontsize" : 1.0}
plt.pie(sizes, labels = labels, colors = colors, autopct ='%1.0f%%')
plt.title('Recruitment Status', fontsize = 20)
plt.axis('equal')
plt.show()
```

```python
# Method 2 - Plot the data using bar chart method

sizes = [2805, 1025, 1004, 526, 181, 107, 74]
labels = 'Recruiting', 'Completed', 'Not yet recruiting','Active, not recruiting', 'Enrolling by invitation','Withdrawn', 'Terminated'
colors = ("orange", "cyan", "brown", "grey", "indigo", "yellow", "red")

plt.bar(labels, sizes, width= 0.5, color = colors, edgecolor="red")
plt.rcParams["figure.figsize"] = (15,11.5)

plt.title("Recruitment Status", fontsize = 15)
plt.xlabel("Status of Study", fontsize = 15)
plt.ylabel("Count", fontsize = 15)
plt.show()
```

```python
# Method 3 - Plot the data using histogram method

x = [74, 181, 1025, 2085, 1004, 526, 107, 74]
y = [0, 500, 1000, 1500, 2100]
plt.hist(x,y)

plt.title("Recruitment Status")
plt.xlabel("Status of Study")
plt.xticks(y)
plt.ylabel("Count")

plt.show()
```

```python
Among the three Data Visualisation methods we've used here, the Bar Plot one is the most preferred one as it shows the exact value of data.
    Unlike Pie chart, which shows the percentage of the label among the total data, and
    Histogram which shows the data as a continuous function,
both of which are not suitable for the dataset that we've been given here.
```
