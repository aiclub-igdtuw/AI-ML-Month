* Plot pairplot for Region Wise Data Distribution of Parameters: sales, postal code, discount,
profit and quantity.

```python
sns.pairplot(d1, vars=['Sales','Postal Code'], hue='City')
```

* Is there any Impact of Sales through Ship Mode







Yes, for different shipping modes , their are variation in Sales.

```python
sns.pairplot(d1, x_vars=['Sales'], y_vars=['Ship Mode'], hue='Ship Mode')
```

```python
plt.bar(d1['Ship Mode'] , d1['Sales'])
plt.xlabel("Ship Mode")
plt.ylabel("Sales")
plt.title("Impact of Sales through Ship Mode")
plt.show()
```

* Find Segment Wise Sales and its Impact. Plot bar plot and find which segment has highest
sales and profit.

```python
plt.bar(d1['Segment'] , d1['Sales'])
plt.xlabel("Segment")
plt.ylabel("Sales")
plt.title("Segment Wise Sales and its Impact")
plt.show()
```

Home Office Segment has the highest sales.

```python
plt.bar(d1['Segment'] , d1['Profit'])
plt.xlabel("Segment")
plt.ylabel("Profit")
plt.title("Profit VS Segment")
plt.show()
```

Corporate Segment has the highest profit.

* Create bar plot for Sub-Category Wise Order Counts.

 Which Sub-Category Wise has maximum order and minimum order?

```python
plt.bar( d1['Quantity'], d1['Sub-Category'])
plt.ylabel("Sub-Category")
plt.xlabel("Order Counts")
plt.title(" Sub-Category Wise Order Counts")
plt.show()
```

The Accessories sub category has the maximum order.

The Storage Sub Category has minimum order.
