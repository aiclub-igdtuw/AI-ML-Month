Q1. There are different methods to replace null values in a data. Mention all of them and mention which one you chose and why.

```python
data.isnull().sum()
```

the different methods to replace null values or rather remove null values in a data are:

1) fill the missing values by finding mean/median/mode(mean and median work for numerical data whereas mode works for categorical data)

2) Drop the null values

3) replace the missing value(categorical) by a new value 

```python
a=data["Acronym"].mode()
```

* acronym has few null values and is categorical data.
* we take mode of the values and replace null values by the mode

```python
a
```

```python
data["Acronym"].replace(np.NaN,"COVID-19",inplace=True)
```

* Gender also has few missing values and can be replaced by the mode of the data. 

```python
data["Gender"].unique()
```

```python
m=data["Gender"].mode()
```

```python
m
```

```python
data["Gender"].replace(np.NaN,"All",inplace=True)
```

* interventions column has a quite lot of missing values which cannot be dropped.
* the values cannot be replaced by taking the mode too.
* so we replace the missing values by a new value i.e. 'Not Known'. 

```python
data["Interventions"].replace(np.NaN,"Not Known",inplace=True)
```

* outcome measures has missing values which cannot be replaced by mode and replacing by a new value would be meaningless. 
* so we drop the rows which have null value in outcome measures.

```python
data=data.dropna(subset=["Outcome Measures"])
```

* phases column has missing values which are replaced by the mode of remaining values.

```python
data["Phases"].unique()
```

```python
data["Phases"].mode()
```

```python
data["Phases"].replace(np.NaN,"Not Applicable",inplace=True)
```

* These columns have few null values and thus the rows which have null values in these columns are removed 

```python
data=data.dropna(subset=["Enrollment"])
```

```python
data=data.dropna(subset=["Study Designs"])
```

```python
data=data.dropna(subset=["Other IDs"])
```

```python
data=data.dropna(subset=["Primary Completion Date"])
```

```python
data=data.dropna(subset=["Completion Date"])
```

* the missing values in these columns are replaced by a new value each.

```python
data["Results First Posted"].replace(np.NaN,"Not Known",inplace=True)
```

```python
data["Locations"].replace(np.NaN,"Not Known",inplace=True)
```

```python
data["Study Documents"].replace(np.NaN,"Not available yet",inplace=True)
```
