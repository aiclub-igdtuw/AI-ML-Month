```python
sizes = [4362, 805, 335, 102, 104, 75]
labels = 'Adult, Older Adult','Child, Adult, Older Adult','Adult', 'Child', 'Child, Adult','Older Adult'
colors = ("orange", "cyan", "brown", "grey", "indigo", "yellow")
plt.bar(labels, sizes, width=0.5, color = colors, edgecolor="red")
plt.rcParams["figure.figsize"] = (18,11.5)
plt.title("AGE BRACKET DATA", fontsize = 15)
plt.xlabel("Value", fontsize = 15)
plt.ylabel("Count", fontsize = 15)
plt.show()
```
