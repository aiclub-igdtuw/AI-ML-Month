```python
#Loading the dataset
loan_data = pd.read_csv('../input/loan-default-prediction/Default_Fin.csv')
```

```python
#Question 1  What is the number of people who will be given a grace period of exactly 15 days?
count=len(loan_data[(loan_data["Defaulted?"]==1) & (loan_data["Annual Salary"]>100000) & (loan_data["Bank Balance"]>5000) & (loan_data["Employed"]!=0)])
print(count)
```

```python
#Question 2 Find the Maximum fine that can be levied by the bank in the period of 15 days
max_balance=loan_data["Bank Balance"][(loan_data["Defaulted?"]==1) & (loan_data["Annual Salary"]>100000) & (loan_data["Bank Balance"]>5000)].max()
max_fine=(2.5*max_balance*15)/100
print(max_fine)
```

```python
#Question 3  Assuming that all the people falling in the category defaulters have not paid their EMI in the grace period, What is the total amount generated by the bank as a part of the fine?
sum=0
for i in loan_data.Index:
 if loan_data.loc[i-1,'Defaulted?']==0:
   sum+=(2.5*loan_data.loc[i-1,'Bank Balance']*15)/100
print(round(sum,4))
```

```python
#Adding a column indicating if the person is given 15 days grace or not
loan_data["Grace Given?"]=((loan_data["Defaulted?"]==1) & (loan_data["Annual Salary"]>100000) & (loan_data["Bank Balance"]>5000)).astype(int)
```

```python
#Adding a column indicating if the person is given 15 days grace or not
loan_data["Grace Given?"]=((loan_data["Defaulted?"]==1) & (loan_data["Annual Salary"]>100000) & (loan_data["Bank Balance"]>5000)).astype(int)
```

```python
#Question 4  Use a suitable plot to show the difference in bank balance of the people who are not eligible for a grace period, and those who are.
#plot of people who are not eligible for grace
loan_data[loan_data["Grace Given?"] == 0].plot(kind = "scatter", x = "Index", y = "Bank Balance", color = "yellow", ax=plt.gca())
#plot of people who are eligible for grace
loan_data[loan_data["Grace Given?"] == 1].plot(kind = "scatter", x = "Index", y = "Bank Balance", color = "red", ax=plt.gca())

plt.show()
```

```python
#Question 5 Find the difference between the median of the bank balance of those people who are eligible for Grace Period and the original value.
Tmedian=loan_data["Bank Balance"].median()
Gmedian=loan_data["Bank Balance"][(loan_data["Annual Salary"]>100000) & (loan_data["Bank Balance"]>5000)].median()
difference=Gmedian-Tmedian
print(round(difference,4))
```

```python
#Question 6  Use a suitable graph to show the changes in the annual income of those people who are eligible for grace period and those who are not.
#plot of people who are not eligible for the grace period
non_eligible_people=loan_data.loc[loan_data["Grace Given?"]==0,"Annual Salary"]
non_eligible_people.plot.hist(color="yellow",alpha=0.9)

# plot of people who are eligible for the grace period
eligible_people=loan_data.loc[loan_data["Grace Given?"]==1,"Annual Salary"]
eligible_people.plot.hist(color="red",alpha=0.9)

plt.show()
```
