```python
M = 'Adult, Older Adult','Child, Adult, Older Adult','Adult', 'Child', 'Child, Adult','Older Adult'
All = [4267, 774, 258, 100, 94, 74]
Female = [76, 21, 54, 0, 10, 0]
Male = [0, 0, 23, 0, 0, 0]
x_axis = np.arange(len(M))
colors = ("orange", "brown", "indigo")
plt.bar(x_axis+0.20, All, width=0.2, color = colors, label = "all")
plt.bar(x_axis+0.40, Female, width=0.2, color = colors, label = "Female")
plt.bar(x_axis+0.60, Male, width=0.2, color = colors, label = "Male")
plt.rcParams["figure.figsize"] = (18,11.5)
plt.xlabel("Gender", fontsize = 15)
plt.ylabel("Count", fontsize = 15)
plt.xticks(x_axis,M)
plt.legend()
plt.show()
```
