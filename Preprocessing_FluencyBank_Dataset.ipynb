#Move the clip from
source -> FluencyBank folder folder: EpId File: ClipId
destination -> folder: Type folder: as per type column

```python
for ind in df.index:
  # print(files)
  folder_id=df['EpId'][ind] #folder
  print(folder_id)
  file_id=df['ClipId'][ind] #file
  audio_type=df['Type'][ind]
  print(file_id)
  folder_path = Path('content/[CLIP_DIR]/FluencyBank/{folderid}'.format(folderid=folder_id))
  file_name="FluencyBank_{folderid}_{fileid}.wav".format(folderid=folder_id, fileid=file_id)
  file_path = Path('Type/{audiotype}/{filename}'.format(audiotype=audio_type,filename=file_name))
  # file_name="FluencyBank_"+files['EpId']+"_"+files['ClipId']


  file_path = folder_path / file_name

  if file_path.is_file():
    # File exists
    with file_path.open('r') as file:
        # Read the contents of the file
        contents = file.read()
        # print(contents)# Save the content to the file
        with file_path.open('w') as file:
          file.write(contents)

        print(f"File '{file_path}' has been saved.")
  else:
    # File does not exist
    print(f"File '{file_name}' does not exist in the folder '{folder_path}'.")
```
