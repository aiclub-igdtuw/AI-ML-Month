```python
columns = ['degree_p', 'workex','hsc_b','status','gender','specialisation','degree_t','salary','ssc_p','hsc_p',]
final_df = df[columns]
final_df.head()
```

```python
AA = []
for j in final_df.columns:
    if final_df[j].dtype ==object:
        AA.append(j)
AA
```

```python
#dummy variables
dummy = pd.get_dummies(final_df[AA],drop_first=True)
dummy
```

```python
final_df = pd.concat([final_df.drop(AA,axis=1),dummy],axis=1)
final_df.head()
```

Scaling Standard

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns_to_scale = ['ssc_p','hsc_p','degree_p','salary']
final_df[columns_to_scale] = scaler.fit_transform(final_df[columns_to_scale])
final_df.head()
```

training a model

```python
x = final_df['status_Placed']
y = final_df.drop('status_Placed',axis=1)
```

random forest classifier

```python
from sklearn.model_selection import train_test_split
y_train, y_test, x_train, x_test = train_test_split(y, x, test_size=0.30, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc =  RandomForestClassifier(n_estimators=400)
rfc=rfc.fit(y_train, x_train)
```

```python
predictions = rfc.predict(y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(x_test, predictions))
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(x_test,predictions)
print(accuracy)
```

```python
print(confusion_matrix(x_test, predictions))
```

KnearestNeighbours

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
knn = []
for i in range(1,41):
    knn_classifier =KNeighborsClassifier(n_neighbors= i)
    score =cross_val_score(knn_classifier,y_train,x_train,cv=10)
    knn.append(score.mean())
```

```python
plt.figure(figsize=(15,10))
plt.plot([p for p in range(1,41)],knn,color='blue' )
plt.xticks([p for p in range(1,41)])
plt.title('Accuracy at respective K value')
plt.xlabel('Nearest neighbour Value')
plt.ylabel('Accuracy score')
```

```python
from sklearn.tree import DecisionTreeClassifier as DTC
model = DTC()
model.fit(y_train,x_train)
from sklearn.metrics import accuracy_score
print(accuracy_score(x_test,model.predict(y_test)))
```
